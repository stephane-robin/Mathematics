## Dfinition d'une fonction
Dans le langage Python, une fonction est dclare par l'instruction `def` (pour definition), suivie du nom donn  la fonction, et d'une paire de parenthses qui contient les **arguments** ou **paramtres** ventuellement passs  la fonction lors de son appel. Tout le contenu de la fonction est **indent**. Voici un exemple :

````python
traceUneLigne40()

def traceUneLigne40():
	""" Cette fonction trace une ligne de 40 tirets.
	Argument(s) : aucun
	Valeur de retour : aucune """
	print("-" * 40)
    
````

La cellule, ci-dessus, donne la dfinition de la fonction `traceUneLigne40` mais n'excute pas le code. L'excution du code a lieu lorsque l'on appelle la fonction :

````python
traceUneLigne40()
````

Pour pouvoir utiliser une fonction dans un programme, celle-ci doit avoir t dfinie en amont de son appel.   
Dans la dfinition de la fonction, une rapide explication de son fonctionnement est donne dans le commentaire dlimit par `"""` (qui permet d'crire un commentaire sur plusieurs lignes). Ce commentaire permet de faciliter la lecture du programme aux autres dveloppeurs. On appelle cette partie *docstring*. La commande ```help(nomFonction)```, insre dans votre programme, provoquera son affichage.

````python
help(traceUneLigne40)
````

Enfin, dans la dfinition de la fonction, on trouve bien videmment le code  excuter : `print("-" * 40)`.

### Valeur en retour
Une fonction peut, si on le dsire, retourner une valeur (le rsultat d'un calcul par exemple) afin qu'elle soit accessible dans le programme principal. Pour se faire, on utilise l'instruction `return()`:

````python
# Dfinition de fonctions
def puissance(a,b):
    """ Cette fonction calcule a puissance b.
    Argument(s) : a et b
    Valeur de retour : a^b """
    resultat = a ** b
    return(resultat)    #la fonction retourne la valeur contenue dans la variable resultat

# Programme principal
x = int(input("Nombre  lever au cube : "))
res = puissance(x,3)
print("Rsultat :", res)
````

## Porte des variables
### Variable locale
Les variables cres  l'intrieur d'une fonction n'existent que dans la fonction. Regardez l'exemple ci-dessous : la variable x de la fonction n'est pas la mme que la variable x du programme principal.

````python
def maFonction():
    """
    Fonction qui donne la valeur 10  la variable locale x.
    """
    
    
    x = 10
    return x
    

# Programme principal
x = 3
maFonction()
print(x)
````

La variable x vaut toujours 3 aprs appel de la fonction ! La variable x de la fonction est une **variable locale**  la fonction.  

Mme si la fonction et le programme principal utilisent le mme nom de variable (par exemple x ici), il y a en fait deux variables diffrentes. La variable x qui existe dans la fonction (variable locale  la fonction) et la variable x qui existe dans le programme principal (variable locale au programme principal).

### Variable globale
Les variables locales sont donc modifiables uniquement  l'endroit o elles ont t cres, ce qui n'est parfois pas trs pratique. Par exemple, dans un jeu vido, il est fort probable que le nombre de points de vie du hros (variable `pointsDeVie`) doive tre modifi  de multiples endroits du programme. Une solution serait de passer en argument  chaque fonction la variable `pointsDeVie` et de retourner la nouvelle valeur, mais cela alourdirait considrablement le code.  
Une solution consiste  prciser dans la dfinition de la fonction que la variable `pointsDeVie`  considrer est la variable __*globale*__, ce qui se fait facilement avec l'instruction `global pointsDeVie` :

````python
def coupEpee():
    """
    Fonction qui fait perdre 20 points de vie au hros  la suite d'un coup d'pe.
    """
    global pointsDeVie      # C'est la variable globale pointsDeVie qui est traite dans cette fonction.
    pointsDeVie = pointsDeVie - 20
    if pointsDeVie == 0 :
        print("Votre hros est mort")

# Programme principal
pointsDeVie = 100          # Le hros nait avec 100 pts de vie
coupEpee()
print("Le hros a",pointsDeVie,"points de vie")
````

Le fait de prciser  la fonction `coupEpee()` que la variable  traiter est la variable **globale** l'autorise  la modifier. On remarque que la modification a bien t prise en compte dans le programme principal.


**Attention !** Utiliser des variables globales dans une fonction rend la fonction dpendante du programme principal. C'est dangereux et souvent pas souhaitable ! En effet, on prfrera toujours avoir des fonctions indpendantes du programme principal, fonctions qui pourront facilement tre rutilises dans d'autres programmes.

## Fonctions rcursives
Une fonction peut aussi s'appeler elle-mme ! On parle alors de **fonction rcursive**. Observez le programme ci-dessous. La fonction multiplication s'appelle jusqu' ce que `a` devienne nul. Essayez de suivre l'excution de ce programme pour comprendre le principe des fonctions rcursives, par exemple avec `a=3` et `b=5`.

````python
# Dfinitions
def multiplication(a,b,res):
    if a != 0 :
        return(multiplication(a-1,b,res+b))
    else:
        return(res)

# Programme principal
resultat = multiplication(3,5,0)
print(resultat)
````

La fonction `multiplication` est appele 4 fois dans ce programme :



Attention, lorsque l'on crit des fonctions rcursives, il faut toujours veiller  ce que la fonction ne s'appelle qu'un nombre fini de fois.

__Remarque__ : bien que le code crit  l'aide de fonctions rcursives soit souvent esthtique et concis, cela se fait souvent au dtriment de l'efficacit du programme (augmentation de la complexit).

## Rutilisation des fonctions
Comme mentionn en introduction, lors de dveloppement informatique consquent, le dcoupage de son code en diffrentes fonctions est primordial. En effet, il va faciliter la rutilisation des fonctions dj codes.  
Il est ainsi trs frquent de regrouper la dfinition de toutes ses fonctions dans un mme fichier python (.py) que l'on appelle un *module de fonctions*. Ce fichier contiendra uniquement des dfinitions de fonctions.  
Pour utiliser les fonctions prsentes dans ce module, il suffira de les *importer* dans le fichier principal  l'aide de l'instruction `from <nom_du_module> import <nom_de_la_fonction>` que l'on place tout en haut du fichier.  


Par exemple, si la fonction `puissance()` a t dfinie dans le module (fichier) __*mesFonctions.py*__ , je peux l'importer dans mon fichier principal de la manire suivante :
```python
from mesFonctions import puissance  #  raliser dans les 1res lignes du programme

puissance(4)   # L'appel  la fonction est alors possible !
```
